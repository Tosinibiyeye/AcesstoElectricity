### imputing  value for the missing value in the "Access to electricity (% of population)" and "Access to clean fuels for cooking"

```python
df_combined_global_details= df_combined_global_details.fillna(df_combined_global_details['Access to electricity (% of population)'].mean())
```

```python
df_combined_global_details= df_combined_global_details.fillna(df_combined_global_details['Access to clean fuels for cooking'].mean())
```

```python
df_combined_global_details.head(10)
```

```python
# Reshape the variables into NumPy arrays and put them into separate objects.

X = df_combined_global_details['Access to electricity (% of population)'].values.reshape(-1,1)
y = df_combined_global_details['Access to clean fuels for cooking'].values.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
# Split data into a training set and a test set.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Create a regression object.

regression = LinearRegression()  # This is the regression object, which will be fit onto the training set.
```

```python
# Fit the regression object onto the training set.

regression.fit(X_train, y_train)
```

```python
# Predict the values of y using X.

y_predicted = regression.predict(X_test)
```

```python
# Create a plot that shows the regression line from the model on the test set.

plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('Access to electricity (% of population) vs Access to clean fuels for cooking')
plot_test.xlabel('Access to clean fuels for cooking')
plot_test.ylabel('Access to electricity (% of population)')
plot_test.show()
```

### The Prediction model has two far extremely high populations with no access to electricity and access to clean fuels for cooking 
### Another set of the population with extremely/100% both access to electricity and access to clean fuels for cooking 

#### The primary indication/meaning is that access to electricity varies directly to access to clean fuels for cooking 

```python
# Create objects that contain the model summary statistics.

rmse = mean_squared_error(y_test, y_predicted) # This is the mean squared error
r2 = r2_score(y_test, y_predicted) # This is the R2 score.
```

```python
# Print the model summary statistics. This is where you evaluate the performance of the model.

print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```
