# import World Development Indicators dataframe 

```python
df_WDI_Country = pd.read_csv(join(PATH,'2_1_Original_Data','WDI_Country.csv'))
df_WDI_Country.head(10)
```

```python
df_WDI_Country.shape
```

```python
# names of the columns of dataset details
df_WDI_Country.info()
```

```python
#check for missing values
df_WDI_Country.isnull().sum()
```

```python
# renmae "Table_Name" column,to read Country

df_WDI_Country.rename(columns = {'Table_Name' : 'Country'}, inplace = True)
```

```python
df_WDI_Country.rename(columns = {'Short_Name' : 'Country_Short_Name'}, inplace = True)
df_WDI_Country.rename(columns = {'Long_Name' : 'Country_Long_Name'}, inplace = True)
```

```python
# Drop unnecessary/empty columns 
df_WDI_Country = df_WDI_Country.drop(['Alternative conversion factor'], axis=1)
df_WDI_Country = df_WDI_Country.drop(['PPP survey year'], axis=1)
df_WDI_Country = df_WDI_Country.drop(['Latest water withdrawal data'], axis=1)
```

```python
df_WDI_Country.head()
```

# merge global-data-on-sustainable-energy dataframe with World Development Indicators dataframe

```python
df_combined_global_details = pd.merge(df_global_data, df_WDI_Country, how='left', on='Country')
df_combined_global_details.head()
```

```python
df_combined_global_details.shape
```

```python
# Export 'ords_prods_group' dataframe to prepared data folder in pickle format
df_combined_global_details.to_csv(join(PATH,'2_2_Prepared_Data','df_combined_global_details.csv')) 
```
